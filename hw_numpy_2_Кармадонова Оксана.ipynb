````python

a =  np.array(np.arange(1, 13, 1).reshape(4, 3), dtype='float')
b = np.arange(1, 5, 2).reshape(2,)
a[b] = np.nan

c = np.empty((3,3,))
c[:] = np.nan


def non_nan(a):
    if np.nansum(a) == 0:
        a[np.isnan(a)] = 0
    else:
        a[np.isnan(a)] = np.nanmean(a)
    return a
        

print(a)
print(c)
print(non_nan(a))
print(non_nan(c))
````

````python
#Сдаюсь, игру не знаю, код нашла в сети, надо разбираться

#Вселенная Игры Жизни представляет собой бесконечную двумерную ортогональную сетку квадратных ячеек , 
#каждая из которых находится в одном из двух возможных состояний, 
#живом или мертвом (или заселенном и незаселенным , соответственно).

 # Каждая ячейка взаимодействует со своими восемью соседями , 
#которые являются ячейками, смежными по горизонтали, вертикали или диагонали.

def iterate(Z):
   
    N = (Z[0:-2,0:-2] + Z[0:-2,1:-1] + Z[0:-2,2:] +
         Z[1:-1,0:-2] + Z[1:-1,2:] +
         Z[2:  ,0:-2] + Z[2:  ,1:-1] + Z[2:  ,2:])

# Выживает любая живая клетка с двумя-тремя живыми соседями.
#Любая мертвая клетка с тремя живыми соседями становится живой клеткой.
#Все остальные живые клетки умирают в следующем поколении. Точно так же все другие мертвые клетки остаются мертвыми.
    
    birth = (N == 3) & (Z[1:-1,1:-1]==0)
    survive = ((N == 2) | (N == 3)) & (Z[1:-1,1:-1] == 1)
    Z[...] = 0
    Z[1:-1,1:-1][birth | survive] = 1
    return Z

Z = np.random.randint(0,2,(50,50))
for i in range(100):
    print(Z)
    Z = iterate(Z)
````
